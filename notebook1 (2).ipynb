```python
import requests

address_cache = {}

API_KEY = '4550892d54c44c8a961fa7b848ba9929'

def get_lat_long(address):
    if address in address_cache:
        cached_result = address_cache[address]
        if cached_result:
            return cached_result
        else:
            return None

    url = 'https://api.opencagedata.com/geocode/v1/json'
    params = {
        'q': address,
        'key': API_KEY,
        'limit': 1,
        'no_annotations': 1
    }

    response = requests.get(url, params=params)
    data = response.json()

    if data['total_results'] > 0:
        location = data['results'][0]['geometry']
        lat, lng = location['lat'], location['lng']
        address_cache[address] = (lat, lng)
        return lat, lng
    else:
        print("Error: No results found")
        address_cache[address] = None

# Example usage
# address = 'Shivajinagar Police Station, Bengaluru, Karnataka, India'
# latitude, longitude = get_lat_long(address)
# print(f"Latitude: {latitude}, Longitude: {longitude}")

i=0
for index, row in accident.iterrows():
    address = f"{row['UNITNAME']}, {row['DISTRICTNAME']}, Karnataka, India"
    lat_lon = get_lat_long(address)
    if lat_lon:
        row['Latitude'], row['Longitude'] = lat_lon
    print(i)
    i = i+1

print(accident)
```

```python
grouped1 = accident_map.groupby(['UNITNAME', 'DISTRICTNAME', 'Accident_Spot', 'Accident_Road']).size().reset_index(name='count')

result_dict = {}

i = 0
for _, row in grouped1.iterrows():
    unit = row['UNITNAME']
    district = row['DISTRICTNAME']
    spot = row['Accident_Spot']
    road = row['Accident_Road']
    count = row['count']

    # Create key if it doesn't exist
    if (unit, district) not in result_dict:
        result_dict[(unit, district)] = {}

    # Create nested dictionary if it doesn't exist
    if (spot, road) not in result_dict[(unit, district)]:
        result_dict[(unit, district)][(spot, road)] = count
    else:
        result_dict[(unit, district)][(spot, road)] += count
    print(i)
    i+=1
print(result_dict)
```

```python
severity_to_score = {
    'Grievous Injury': 100,
    'Fatal': 80,
    'Simple Injury': 60,
    'Damage Only': 40
}

accident_map['Score'] = accident_map['Severity'].map(severity_to_score).fillna(0)

grouped = accident_map.groupby(['UNITNAME', 'DISTRICTNAME']).agg({'Score': 'sum', 'Severity': 'count'}).reset_index()

grouped['Normalized_Score'] = grouped['Score'] / grouped['Severity']

grouped['final_score'] = grouped['Normalized_Score'].apply(lambda x: x if x > 50 else 50)
```

```python
final_scores_dict = {}

for index, row in grouped.iterrows():
    district = row['DISTRICTNAME']
    unit = row['UNITNAME']
    final_score = row['final_score']

    final_scores_dict[(district, unit)] = final_score

print(final_scores_dict)
```

```python
import folium
import re

karnataka_map = folium.Map(location=[12.9716, 77.5946], zoom_start=7)

def get_marker_color(severity_score):
    if severity_score > 75:
        return 'red'
    elif severity_score > 60:
        return 'orange'
    else:
        return 'blue'

i=0
visited_pairs = set()

for index, row in accident_map.iterrows():
    unitname_parts = row['UNITNAME']
    unit = ' '.join(re.split(r'\s+', unitname_parts)[:-2])
    district = row['DISTRICTNAME']

    if (unit, district) in visited_pairs:
        continue  # Skip processing if already visited

    visited_pairs.add((unit, district))

    severity_score = final_scores_dict.get((district, unitname_parts), 50)

    unit_count = district_unit_counts[(district, unitname_parts)]
    unit_count_2023 = district_unit_counts_2023.get((district, unitname_parts), 0)

    subset = accident_map[(accident_map['UNITNAME'] == unitname_parts) & (accident_map['DISTRICTNAME'] == district)]

    if not subset.empty:
        first_road = subset['Accident_Road'].iloc[0]
    else:
        first_road = 'Unknown'

    # HTML formatting for popup
    popup_text = f"""
    <div style="font-family: Arial, sans-serif; color: black;">
        <h4 style="margin-bottom: 5px;">{unit}, {district}</h4>
        <p>Severity Score: {severity_score:.2f}</p>
        <p>Accident Count(7 years): {unit_count}</p>
        <p>Accident Count 2023: {unit_count_2023}</p>
        <p>Unsafe Road: {first_road}</p>
    </div>
    """

    marker_color = get_marker_color(severity_score)

    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color=marker_color, icon='info-sign')
    ).add_to(karnataka_map)

    print(i)
    i += 1

karnataka_map.save('karnataka_map.html')
karnataka_map
```
